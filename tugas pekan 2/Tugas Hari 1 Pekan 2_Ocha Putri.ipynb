<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab:
- Perbedaan Series dengan list dan dictionary : 
1. Series memiliki index label yang bisa ditentukan sendiri sedangkan list dan dictionary tidak memiliki index label secara default. 
2. Series memungkinkan untuk menyimpan data dengan tipe yang berbeda-beda sementara list dan dictionary hanya bisa menyimpan data dengan tipe yang sama.
3. Series memiliki banyak fungsi dan methos yang lebih canggih dibandingkan list dan dictionary, seperti statistik deskriptif, pengelompokan data, dan manipulasi data. 
4. Series lebih cepat dalam melakukan dibandingkan list dan dictionary, terutama data yang besar



- Perbedaan dataframe dengan series : DataFrame adalah tipe data 2-dimensi yang memiliki baris dan kolom dengan index label pada setiap baris dan kolom. Sedangkan Series adalah tipe data 1-dimensi hanya terdiri dari satu kolom dengan index label pada setiap baris. Keduanya memiliki fungsi dan method yang berbeda, tetapi DataFrame lebih fleksibel dalam menyimpan dan mengolah data dengan tipe yang berbeda-beda, sedangkan Series lebih cepat dalam melakukan operasi. 

```python
# code here
x = pd.Series([6,3,4,6], index=['a', 'b', 'c', 'd'])
print(x)
```

Deskriptif statistik adalah cabang statistik yang bertujuan untuk mengumpulkan, mengorganisasi, menganalisis, dan menjelaskan data numerik dengan menggunakan metode seperti visualisasi, tabulasi, dan ukuran-ukuran seperti rata-rata, standar deviasi, dan frekuensi untuk memahami karakteristik dan pola data. <h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
df_data.describe()
```

```python
mean_price = df_data["price"].mean()
print("rata-rata price:", mean_price)
```

```python
std_depth = df_data["depth"].std()
print("Standar deviasi depth:", std_depth)
```

```python
max_carat = df_data["carat"].max()
print("Maximum carat:", max_carat)
```
