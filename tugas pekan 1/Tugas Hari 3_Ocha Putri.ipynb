<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

//isi jawaban text disini

**Built-In Function** ialah sebutan bagi fungsi yang sudah bawaan dari bahasa pemrograman itu sendiri

3 built-in Function di python:
1. bin() digunakan untuk mengubah integer dan mengembalikan nilai dalam bentuk string biner
2. bool() merupakan salah satu fungsi yang mengubah sebuah nilai menjadi nilai Boolean (True atau False) dan mengembalikannya.
3. pow() berfungsi untuk mendapatkan nilai pangkat dari suatu bilangan



```python
# Contoh penggunaan built in function

# contoh bin()
number = 10
print(f'Biner dari {int(number)} adalah ', bin(number))

# contoh bool()
test = 0
print(f'{int(test)} adalah ', bool(test))

# contoh pow()
print(pow(3,4)) # 3 pangkat 4
```

# Soal 7. Mapping Fungsi lambda

Hitunglah jumlah bilangan genap dari sebuah objek list berikut menggunakan lambda function

```python
obj_list = [i for i in range(1000)]
hasil = list(filter(lambda a : a % 2 == 0, obj_list))


print(len(hasil))
```

Expected output:

500
